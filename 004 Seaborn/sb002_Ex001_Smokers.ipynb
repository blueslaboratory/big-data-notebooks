```python
# path = "../datos"
path = "/content/drive/My Drive/Clara del Rey/BIG DATA/Programacion/FTP/000 Datos/"
datos = pd.read_csv(path + "insurance.csv", sep = ",")
datos.head()
```

```python
# devuelve un FacetGrid
g = sns.relplot(
    datos,
    x="age",
    y="charges",
    size = "bmi",
    hue="sex",
    style="smoker"
)


# con fig se accede al objeto FIgure de matplotlib
g.fig.suptitle("Conjunto de datos de Insurance")
g.fig.subplots_adjust(top=.85)
```

E001. b) Haz una versión del ejercicios anterior con cuatro subplots, uno por cada región.

```python
# devuelve un FacetGrid
g = sns.relplot(
    datos,
    x="age",
    y="charges",
    size="bmi",
    hue="sex",
    style="smoker",
    col="region",
    col_wrap=2
)


# con fig se accede al objeto FIgure de matplotlib
g.fig.suptitle("Conjunto de datos de Insurance")
g.fig.subplots_adjust(top=.90)
```
