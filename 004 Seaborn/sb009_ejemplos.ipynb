```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(
    data=tips,
    x="total_bill"
)
```

```python
sns.catplot(
    data=tips,
    x="total_bill",
    y="day"
)
```

Con los datos de pinguinos, muestra un gráfico con un diagrama de violín de la longitud del pico.
Tiene que tener un violín por cada especie. Además dentro de cada subplot, hay que diferenciar entre machos y hembras.

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.violinplot(
    penguins,
    x = 'species',
    y = 'bill_length_mm',
    hue = 'sex',
    split = True
)
```

Con catplot()

```python
sns.catplot(
    penguins,
    x = 'species',
    y = 'bill_length_mm',
    hue = 'sex',
    split = True,
    kind = "violin"
)
```
