```python
# 05/11/2024
import seaborn as sns
import matplotlib.pyplot as plt

# %matplotlib qt
%matplotlib inline
```

# Histogramas

```python
iris = sns.load_dataset("iris")
iris.head()
```

En seaborn podemos hacer histogramas con la función a nivel de figura displot() o la función a nivel de axes histpplot().

https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

## Función displot()

```python
# El histograma representa el area bajo la curva (kde=True, stat="proportion")
sns.displot(
    data=iris,
    x="sepal_length",
    hue="species",
    col="species",
    kde=True,
    stat="proportion"
)
```

```python
sns.displot(
    data=iris,
    x="sepal_length",
    hue="species",
    col="species",
    cumulative="True"
)
```

Las funciones a nivel de figura admiten parámetros de las de *axes*, pero hay que buscarlos en la documentación correspondiente. En este ejemplo, bins es un argumento para histplot().

```python
sns.displot(
    data=iris,
    x="sepal_length",
    hue="species",
    col="species",
    bins=50
)
```

```python
sns.displot(
    iris,
    x = "sepal_length",
    hue = "species",
    kde=True,
    stat="proportion"
)
```

Usando el parámetro y en lugar de x:

```python
sns.displot(
    data=iris,
    y="sepal_length",
    hue="species",
    col="species"
)
```

```python
sns.displot(
    iris,
    y = "sepal_length",
    hue = "species"
)
```

Usando los dos se obtiene un diagrama en dos dimensiones.

```python
g = sns.displot(
    iris,
    x = "sepal_width",
    y = "sepal_length",
    hue = "species"
)

g.legend.set_title(title="Especies")
plt.figure()
```

## Función histplot()

Con la función a nivel de *axes*:

```python
sns.histplot(
    data=iris,
    x="sepal_length",
    stat = "proportion",
    hue = "species",
    multiple = "fill"
)
```
