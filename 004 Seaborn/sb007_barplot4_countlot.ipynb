```python
# 07/11/2024
# https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

La función barplot queivale a catplot con kind = 'bar'.
Agrupa por una columna categórica y hace un cálculo agregado por otra columna, por defecto la media.

```python
ax= sns.barplot(
    iris,
    x ="species",
    y = "petal_length"
)

ax.bar_label(ax.containers[0], fontsize=10)
```

Para probar más opciones cargamos el fichero de pingüinos

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.barplot(
    penguins,
    x ="species",
    y = "bill_length_mm",
    hue = 'sex'
)
```

```python
sns.catplot(
    penguins,
    x ="species",
    y = "bill_length_mm",
    hue = 'sex',
    col = 'island',
    kind = 'bar'
)
```

Probamos otro estimador

```python
sns.barplot(
    penguins,
    x ="species",
    y = "bill_length_mm",
    hue = 'sex',
    estimator = 'std'
)
```

Si solo queremos contar, tenemos countplot() o catplot() con kind = 'count'

```python
sns.countplot(
    penguins,
    x = 'species',
    hue = 'sex'
)
```

```python
sns.catplot(
    penguins,
    x = 'species',
    hue = 'sex',
    kind = 'count',
    col = 'island'
)
```
