```python
datos = sns.load_dataset("titanic")
datos.head()
```

```python
primera = datos[datos.pclass == 1]
print(primera.fare.max())

segunda = datos[datos.pclass == 2]
print(segunda.fare.max())

tercera = datos[datos.pclass == 3]
print(tercera.fare.max())
```

```python
sns.displot(
    datos,
    x = "fare",
    bins = 20
)
```

```python
sns.displot(
    datos,
    x = "fare",
    bins = 80,
    hue = "class",
    col = "class",
    col_wrap = 2,
    facet_kws = {"sharey":False, "sharex":False}
)
```

```python
g = sns.displot(
    datos,
    x = "fare",
    bins = 50,
    hue = "class",
    col = "class",
    col_wrap = 2,
    facet_kws = {"sharey":False, "sharex": False}
)

# avanzado
g.fig.axes[1].set_xlim(0,100)
g.fig.axes[2].set_xlim(0,100)
```
