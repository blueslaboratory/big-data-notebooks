```python
# UPDATED: 14/11/2024

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = sns.load_dataset("penguins")
datos.head()
```

Un pairplot con rectas de regresiones sin más... Mejorable

```python
# para los histogramas no sirven las medidas del resto de gráficos,
# las escalas son para los scatter, no para los histogramas

# sns.pairplot(data = datos, kind = "reg")
sns.pairplot(
    data = datos,
    kind = "reg",
    corner = True
)
```

```python
# Crea la cuadrícula de subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Genera cada gráfico en un subplot específico
sns.histplot(data=datos, x="bill_length_mm", ax=axes[0])
sns.histplot(data=datos, x="bill_depth_mm", ax=axes[1])
sns.histplot(data=datos, x="flipper_length_mm", ax=axes[2])
sns.histplot(data=datos, x="body_mass_g", ax=axes[3])

# Ajusta el espaciado entre los gráficos
plt.show()
```

Separando por especies

```python
sns.pairplot(data = datos, hue = "species", kind = "reg")
```

```python
sns.pairplot(data = datos, kind = "hist")
```

```python
sns.pairplot(data = datos, kind = "kde")
```

Vamos a seleccionas las columnas que se analizan

```python
sns.pairplot(
    data = datos,
    hue = "species",
    x_vars="body_mass_g",
    y_vars=["bill_length_mm", "bill_depth_mm"],
    kind = "reg"
)
```

Ahora vamos a centrarnos en la profundidad del pico y el peso (body_mass_g).

Veamos si hay relación lineal separando por especie y sexo/isla.

Empezamos por peso contra profundidad del pico sin separar.

Cuanto más fina es la sombra(intervalo de confianza) de la recta indica que es más exacta

```python
sns.lmplot(
    data = datos,
    x = "bill_depth_mm",
    y = "body_mass_g"
)
```

Hay que separar por especies

```python
sns.lmplot(
    data = datos,
    x = "bill_depth_mm",
    y = "body_mass_g",
    hue = "species"
)
```

Veamos si la isla cambia las cosas

```python
sns.lmplot(
    data = datos,
    x = "bill_depth_mm",
    y = "body_mass_g",
    hue = "species",
    col = "island",
    col_wrap = 2
)
```

También separando por sexo

```python
sns.lmplot(
    data = datos,
    x = "bill_depth_mm",
    y = "body_mass_g",
    hue = "species",
    col = "island",
    row = "sex"
)
```
