Función interact()

```python
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
def funcion(x):
   return x*3
```

```python
# [-1*10, 3*10]
interact(funcion, x = 10)
```

Según el segundo argumento, se crea un control u otro

```python
# x = True (1) --> 3*1
# x = True (0) --> 3*0
interact(funcion, x = True)
```

Se puede usar un decorador. Se crea un control por cada argumento de interact, el tipo de control depende de los valores que se asignen al argumento.


```python
# te crea el slider:
# interact(doble, x = 1)
@interact(x = 1)
def doble (x):
   return x*2
```

En el siguiente ejemplo, además se usa más de un control

```python
@interact(x=True, y=1.0)
def g(x, y):
    return (x, y)
```

```python
@interact(x=True, color =['rojo', 'verde'])
def g(x, color):
    return (x, color)
```

```python
@interact(x=True, color =[('rojo',7), ('verde',55)])
def g(x, color):
    return (x, color)
```

Para pasar más parámetros a los widgets, hay que usar la clase correspondiente

```python
@interact(x=widgets.Combobox(options=["Chicago", "New York", "Washington"], value="Chicago"))
def g(x):
    return (x)
```

```python
@interact(x=widgets.Dropdown(options=[("Chicago", 23), ("New York", 7), ("Washington", 55)], value=55), description = "Pruebas")
def g(x):
    return (x)
```

Ejemplo con el conjunto de datos iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = sns.load_dataset('iris')

@interact(specie = ['setosa', 'versicolor', 'virginica'])
def cargarHistograma(specie):
    parcial = datos[datos.species == specie]
    sns.displot(parcial.petal_length)
    plt.show()
```

Ejercicio: Añadir una checkbox para mostrar histograma o kde.

```python
datos = sns.load_dataset('iris')

@interact(kde=True, specie = ['setosa', 'versicolor', 'virginica'])
def cargarHistograma(kde, specie):

    parcial = datos[datos.species == specie]

    if kde:
        sns.kdeplot(parcial.petal_length)
    else:
        sns.displot(parcial.petal_length)

    plt.show()
```
