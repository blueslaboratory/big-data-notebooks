En seaborn hay funciones a nivel de figura y a nivel de *axes*. Las funciones a nivel de figura agrupan varios tipos de gráficos similares. Por ejemplo, la función a nivel de figura relplot() incluye gŕaficos de scatter con la función a nivel de *axes* scatterplot() o de línea con lineplot().
image.png


La principal utilidad de las funciones a nivel de figura es que hacen gráficos con varios subplots según el valor de una columna.

Las funciones a nivel de figura devuelven un objeto FacetGrid, a partir del cual se puede acceder al objeto Figure de matplotllib.

Las funciones a nivel de *axes* solo crean un subplot. Devuelven un objeto Axes.

En los dos casos se pueden mostrar los datos con diferentes colores, tamaños o marcadores en función del valor de una columna.

Las funciones a nivel de figura crean los gŕaficos usando una de las funciones a nivel de *axes* asociada, que se indica con el parámeto *kind*. Para relplot() se pueden usar los valores 'scatter' (por defecto) o 'line'.

# Gráficos de scatter

Cargamos los datos de iris:

```python
iris = sns.load_dataset("iris")
iris.head()
```

Uso básico de relplot
https://seaborn.pydata.org/generated/seaborn.relplot.html



```python
sns.relplot(
    iris,
    kind = 'scatter',
    x = "petal_length",
    y = "petal_width"
)
```

Para hacer varios subplots, se usan col, row y col_wrap.
Por ejemplo, el gráfico anterior separando por especie.

```python
sns.relplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    col = "species"
)
```

Ahora en filas y columnas

```python
sns.relplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    col = "species",
    col_wrap = 2
)
```

También hay parámetros para que el color (hue), el tamaño y el marcador (hue, size y style) dependan del valor de una columna.

```python
sns.relplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    hue = "species",
    style = "species"
)
```

Se pueden usar en combinación con col.

```python
sns.relplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    col = "species",
    col_wrap = 2,
    hue = "species",
    style = "species"
)
```

Es habitual tener que usar directamente los objetos de matplotlib. Hay que almacenar el valor que devuelve la función.

```python
# devuelve un FacetGrid
g = sns.relplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    col = "species",
    hue = "species",
    style = "species"
)

# con fig se accede al objeto FIGure de matplotlib
g.fig.suptitle("Conjunto de datos Iris")
g.fig.subplots_adjust(top=.85)
```

Para un solo subplot también se puede usar la función a nivel de *axes* scatterplot().
https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
sns.scatterplot(
    iris,
    x = "petal_length",
    y = "petal_width",
    hue = "species",
    style = "species"
)
```
