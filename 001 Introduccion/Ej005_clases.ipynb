```python
class Persona:

  def __init__(self, nombre, edad, telefonos, correo1, correo2):
    self.nombre = nombre
    self.edad = edad
    self.telefonos = telefonos

    self.correos = dict()
    self.correos['personal'] = correo1
    self.correos['trabajo'] = correo2


  # El __funcion__ se puede utilizar con cualquiera de estas funciones:
  # https://docs.python.org/3/library/functions.html

  def __str__(self):
    return self.nombre

  def __len__(self):
    return len(self.nombre)

  def __getitem__(self, key):
    # return self.telefonos[key]
    return self.correos[key]
```

```python
p = Persona("Ana", 20, ["222", "333"], "personal@correo.com", "trabajo@correo.com")

print(p)

# el print comentado es para el __getitem__: correos
# print(p[1])
print(p['personal'])


print(len(p))
```

Clase para representar fracciones

```python
class Fraccion():

  def __init__(self, numerador, denominador):
    self.numerador = numerador
    self.denominador = denominador

  def __str__(self):
    return str(self.numerador) + "/" + str(self.denominador)

  def sumar(self, f2):
    resul = Fraccion(1, 1)
    resul.numerador = self.numerador * f2.denominador +  f2.numerador * self.denominador
    resul.denominador = self.denominador * f2.denominador
    return resul
```

Probarla

```python
f1 = Fraccion(2, 3)
f2 = Fraccion(1, 3)

r = f1.sumar(f2)

print(r)
```
