```python
import math
class Punto:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return "(" + str(self.x) + "," + str(self.y) + ")"

    def distancia(self, otro):
        return math.sqrt((self.x - otro.x)**2 + (self.y - otro.y)**2)
```

```python
p1 = Punto(2,2)
p2 = Punto(3,3)

print(p1)
print(p2)

print (p1.distancia(p2))
```
