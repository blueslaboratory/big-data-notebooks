```python
# Fibonacci

a, b = 0, 1

while a < 10:

    print(a)

    a, b = b, a+b
```

# Diccionarios

Es una colección de elementos key: value

https://docs.python.org/3/tutorial/datastructures.html#dictionaries

```python
d = {'jack': 4098, 'sjoerd': 4127}
print(d['jack'])

d['jack'] = 2048
print(d['jack'])

```

```python
[('jack', 4098), ('sjoerd', 4127)]
```

```python
print(dict([('jack', 4098), ('sjoerd', 4127)]))
```

Tambien existe "dict comprehension":

```python
{x: x**2 for x in range(10)}
```

```python
cuadrados = {'pos' + str(x) : x ** 2 for x in range(10)}

cuadrados
```

```python
for k, v in cuadrados.items():
    print(k, v)
```

```python
knights = {'gallahad': 'the pure', 'robin': 'the brave'}

for k, v in knights.items():
    print(k, v)
```
