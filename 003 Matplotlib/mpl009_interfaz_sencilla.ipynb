```python
import pandas as pd
```

**Gráficos con pyplot**



```python
import matplotlib.pyplot as plt

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
```

Empezamos con la función x²

```python
x = pd.Series(range(-6,6))
x
```

```python
y = x * x
y
```

En general todas los tipos de gráfico disponibles en los objetos Axes están también en plt.
Plot básico:

```python
plt.plot(x, y)
plt.show()
```

Con más parámetros

```python
plt.plot(x, y, 'x')
plt.ylabel('x al cuadrado')
plt.show()
```

Usando un dataframe

```python
datos = pd.DataFrame()

datos["valores"] = x
datos["cuadrado"] = y

plt.plot("valores", "cuadrado", data = datos)
#plt.plot(datos.valores, datos.cuadrado)

plt.show()
```

Ahora vamos a poner en la misma gráfica a x³.
Primero una gráfica básica, con las dos series en la misma llamada a plot.

```python
cubo  = x**3

plt.plot(x, y, 'rx', x, cubo, 'go')

plt.show()
```

Ahora con más parámetros y con más llamadas a plot

```python
plt.plot(x, y, 'rx', label = 'cuadrado' )
plt.plot(x, cubo, 'go', label = 'cubo')

plt.xlim(-10,10)
plt.ylim(-50,50)

plt.legend()

plt.show()

# path = "/content/drive/My Drive/programacionBigData/notebooks/datos/"
# path = "../datos/"
# plt.savefig(path + "funciones.png")
```

Desde un dataframe

```python
funciones = pd.DataFrame()
funciones["cuadrado"] = y
funciones["cubo"] = cubo

plt.plot(x, funciones)

plt.show()
```

**Subplots**

En esta interfaz se usa el concepto de "subplot actual".
Usando subplot se cambia el subplot actual. Las llamadas a métodos de plt tienen efecto sobre el subplot actual.

Ejemplo con dos subplots, cada uno con su leyenda.

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, y, "r-", label = "cuadrado")
plt.legend()

plt.subplot(212)
plt.plot(x, cubo, "g-", label = "cubo")
plt.legend(loc = 'upper right')
plt.show()
```

Como el anterior, pero con una sola leyenda de figura

```python
fig = plt.figure()

plt.subplot(2,1,1)
plt.plot(x, y, "r-", label = "cuadrado")

plt.subplot(212)
plt.plot(x, cubo, "g-", label = "cubo")

fig.legend(loc = 'upper right')
plt.show()
```

Con otra opción para la leyenda

```python
fig = plt.figure()

plt.subplot(2,1,1)
l1 = plt.plot(x, y, "r-")

plt.subplot(212)
l2 = plt.plot(x, cubo, "g-")

fig.legend((l1[0],l2[0]), ("etiqueta1", "etiqueta2"),loc='upper right')
plt.show()
```
