Parece que hay que usar más de 10 bins (el valor por defecto) para que se vea bien, porque si no la mayoría de los casos estarían en la primera barra.

```python
fig, ax = plt.subplots()

# separamos por billete
categoria_1 = datos[datos.Pclass == 1]
categoria_2 = datos[datos.Pclass == 2]
categoria_3 = datos[datos.Pclass == 3]

ax.boxplot((categoria_1.Fare,
            categoria_2.Fare,
            categoria_3.Fare),
            labels = ["categoria 1", "categoria 2" , "categoria 3"])

ax.set_title("Diagrama de cajas para el precio del ticket en f(categoria)")

ax.set_xlabel("Categoria")
ax.set_ylabel("Precio del ticket")

plt.show()
```

E010. A partir de los datos del Titanic, muestra un  gráfico de violín con el precio del billete, diferenciando por la categoría.

```python
fig, ax = plt.subplots()

# separamos por billete
categoria_1 = datos[datos.Pclass == 1]
categoria_2 = datos[datos.Pclass == 2]
categoria_3 = datos[datos.Pclass == 3]

# para mostrar los cuantiles se pasa una lista de listas. Para cada serie, una lista con los cuantiles que se quieren mostrar
ax.violinplot((categoria_1["Fare"],
               categoria_2["Fare"],
               categoria_3["Fare"]),
               showmedians = True, quantiles = [[0.25, 0.75], [0.25, 0.75], [0.25, 0.75]])

ax.set_title("Diagrama de violin para el precio del ticket en f(categoria)")

ax.set_xlabel("Categoria")
ax.set_ylabel("Precio del ticket")

ax.set_xticks([1, 2, 3], labels = ["categoria 1", "categoria 2" , "categoria 3"])

plt.show()
```
