```python
# path= "../datos/"
path = "/content/drive/My Drive/Clara del Rey/BIG DATA/Programacion/FTP/000 Datos/"
datos = pd.read_csv(path + "iris.csv", sep = ",")
datos.head()
```

Vamos a hacer un diagrama de cajas con la longitud del pétalo, una caja por especie.

```python
fig, ax = plt.subplots()

# separamos por variedad
setosa = datos[datos.variety == "Setosa"]
virginica = datos[datos.variety == "Virginica"]
versicolor = datos[datos.variety == "Versicolor"]

ax.boxplot((setosa["petal.length"],
            virginica["petal.length"],
            versicolor["petal.length"]),
            labels = ["Setosa", "Virginica" , "Versicolor"])

ax.set_title("Diagrama de cajas para la longitud del pétalo para Iris")

ax.set_xlabel("Variedad")
ax.set_ylabel("Longitud del pétalo (cm)")

plt.show()
```

Ahora hacemos el diagrama de violín equivalente.

```python
fig, ax = plt.subplots()

# separamos por variedad
setosa = datos[datos.variety == "Setosa"]
virginica = datos[datos.variety == "Virginica"]
versicolor = datos[datos.variety == "Versicolor"]

# para mostrar los cuantiles se pasa una lista de listas. Para cada serie, una lista con los cuantiles que se quieren mostrar
ax.violinplot((setosa["petal.length"],
               virginica["petal.length"],
               versicolor["petal.length"]),
               showmedians = True, quantiles = [[0.25, 0.75], [0.25, 0.75], [0.25, 0.75]])

ax.set_title("Diagrama de violin para la longitud del pétalo para Iris")
ax.set_xlabel("Variedad")
ax.set_ylabel("Longitud del pétalo (cm)")

ax.set_xticks([1, 2, 3], labels=["Setosa","Virginica", "Versicolor"])

plt.show()
```

Mostrar un histograma con la funcion de densidad

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.violinplot.html#matplotlib.axes.Axes.violinplot

fig, ax = plt.subplots()

# separamos por variedad
ax.violinplot(datos["petal.length"], positions = [0], vert = False);
ax.hist(datos["petal.length"], weights= [1. / len(datos)] * len(datos))

ax.set_title("Diagrama de cajas para la longitud del pétalo para Iris")
ax.set_xlabel("Variedad")
ax.set_ylabel("Longitud del pétalo (cm)")

ax.set_yticks([0])

plt.show()
```
