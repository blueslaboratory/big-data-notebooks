# Función bar()

Diagrama de barras con las clases de billete del Titanic.

Contamos el número de pasajeros de cada clase:

```python
num = titanic.groupby("Pclass")["PassengerId"].count()
num
```

```python
fig, ax = plt.subplots()

# Es necesario guardarlo en una variable para aplicarlo después
b = ax.bar(num.index, num, tick_label = ["Primera", "Segunda", "Tercera"], label = num, color = ["yellow", "blue", "green"])
ax.bar_label(b)

plt.show()
```

Ahora hombres y mujeres por separado:

*  https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html
* https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py





```python
hombres = titanic[titanic.Sex == "male"]
mujeres = titanic[titanic.Sex == "female"]

grHombres = hombres.groupby("Pclass")["PassengerId"].count()
grMujeres = mujeres.groupby("Pclass")["PassengerId"].count()

ancho = 0.4 # el ancho por defecto es 0.8, esto es una especie de ajuste a mano para el gráfico
fig, ax = plt.subplots()

ax.bar(grHombres.index - ancho /2, grHombres,  ancho, tick_label = ["Primera", "Segunda", "Tercera"], label = "Hombres")
ax.bar(grMujeres.index + ancho /2,  grMujeres, ancho, tick_label = ["Primera", "Segunda", "Tercera"], label = "Mujeres")

ax.set_xticks([1, 2, 3], labels=["Primera", "Segunda", "Tercera"])

ax.legend()

plt.show()
```
