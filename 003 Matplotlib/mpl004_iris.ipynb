```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# para google colab:
# path = "../datos/"
path = "/content/drive/My Drive/Clara del Rey/BIG DATA/Programacion/FTP/000 Datos/"
iris = pd.read_csv(path + "iris.csv", sep = ",")
```

```python
# iris = pd.read_csv("../datos/iris.csv")
iris.head()
```

```python
# para convertir una columna de categorías a números se puede usar esto
# puede ser últil para otros ejercicios
iris["varNumero"]  = iris.variety.astype("category").cat.codes
iris["varNumero"]
```

```python
fig, ax = plt.subplots()

fig.suptitle("Conjunto de datos Iris")
ax.scatter("petal.length", "petal.width", c = "varNumero", data = iris)

ax.set_xlabel("Longitud del pétalo (cm)")
ax.set_ylabel("Anchura del pétalo (cm)")

plt.show()
```

Para que tengan diferentes marcadores

```python
fig, ax = plt.subplots()
fig.suptitle("Figure con diferentes marcadores")

ax.scatter("petal.length", "petal.width", c = "yellow", data = iris[iris.variety=="Setosa"], marker = "^", label = "Setosa" )
ax.scatter("petal.length", "petal.width", c = "red", data = iris[iris.variety=="Virginica"], marker = "o", label = "Virginica" )
ax.scatter("petal.length", "petal.width", c = "green", data = iris[iris.variety=="Versicolor"], marker = "s", label = "Versicolor" )

ax.set_xlabel("Longitud del pétalo (cm)")
ax.set_ylabel("Anchura del pétalo (cm)")
ax.legend()

plt.show()
```

En tres subplots

```python
fig, ax = plt.subplots(3, sharex = True)

fig.suptitle("Figure con diferentes marcadores")
ax[0].scatter("petal.length", "petal.width", c = "yellow", data = iris[iris.variety=="Setosa"], marker = "^", label = "Setosa" )
ax[1].scatter("petal.length", "petal.width", c = "red", data = iris[iris.variety=="Virginica"], marker = "o", label = "Virginica" )
ax[2].scatter("petal.length", "petal.width", c = "green", data = iris[iris.variety=="Versicolor"], marker = "s", label = "Versicolor" )

ax[2].set_xlabel("Longitud del pétalo (cm)")
ax[1].set_ylabel("Anchura del pétalo (cm)")
fig.legend()

plt.show()
```
