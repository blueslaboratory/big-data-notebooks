Ejercicio. Comprueba que las funciones random, randn y normal de numpy.random generan secuencias con la media y varianza/desviación esperadas.

Hay que generar un array de, por ejemplo, 100 números aleatorios, calcular su media y varianza y comprobar que son más o menos los teóricos

```python
import numpy as np
from numpy.random import random, randn, normal, randint
%matplotlib inline
```

```python
# 60000 números enteros entre  1 y 6
a = randint(1, 7, 60000)
print(np.mean(a), np.var(a))
fig, ax  = plt.subplots()
ax.hist(a, bins = 6, weights= [1. / len(a)] * len(a))

ax.set_xlim((0, 7))
plt.show()
```

Pedimos 10000 número aleatorios (distribución uniforme) entre 0 y 1. La media teórica es 0.5, la varianza es (b - a)^2/ 12 (https://es.wikipedia.org/wiki/Distribuci%C3%B3n_uniforme_continua)
Con b = 1, a = 0, var = 1/12


```python
np.random.seed(12)
a = random(10000)
print(np.mean(a), np.var(a), np.std(a))
fig, ax  = plt.subplots()
ax.hist(a, weights= [1. / len(a)] * len(a))
plt.show()
```

Ahora probamos la distribución normal

```python
# se pide una distirbución de media 3 y desviación típica 25
a =  normal(0, 1, 100000)
print(np.mean(a), np.var(a), np.std(a))
fig, ax  = plt.subplots()
ax.hist(a, bins = 100,  weights= [1. / len(a)] * len(a))
plt.show()
```
