E005 (SOL): Para los datos de insurance.csv, crea un gráfico de tarta con el número de hijos de los asegurados. Las categorías serán: 0, 1, 2, 3 y 4 o más.

Empezamos creando una nueva columna como children, pero sustituyendo los valores mayores que cuatro por cuatro.

```python
children2 = [x if x < 4 else 4 for x in datos.children]
datos['children2'] = children2
```

```python
num = datos.children2.value_counts().sort_index()
num
```

```python
fig, ax = plt.subplots()

ax.pie(num, labels = ["0", "1", "2", "3", "4+"], autopct='%d%%')

fig.suptitle("Número de hijos de los asegurados")

plt.show()
```

Ej 6: Crea una figura con cuatro subplots como el del ejercicio anterior, uno por cada región.


```python
labels = ["0", "1", "2", "3", "4+"]

fig, ax = plt.subplots(2, 2)
fig.set_size_inches(w = 8, h = 6)
fig.suptitle("Número de hijos de los asegurados")

numnw = datos[datos.region=='northwest'].children2.value_counts().sort_index()
ax[0, 0].pie(numnw, labels = labels, autopct='%d%%')
ax[0, 0].set_title("Northwest")

numne = datos[datos.region=='northeast'].children2.value_counts().sort_index()
ax[0, 1].pie(numne, labels = labels, autopct='%d%%')
ax[0, 1].set_title("Northeast")

numsw = datos[datos.region=='southwest'].children2.value_counts().sort_index()
ax[1, 0].pie(numsw, labels = labels, autopct='%d%%')
ax[1, 0].set_title("Southwest")

numse = datos[datos.region=='southeast'].children2.value_counts().sort_index()
ax[1, 1].pie(numse, labels = labels, autopct='%d%%')
ax[1, 1].set_title("Southeast")

plt.show()
```
