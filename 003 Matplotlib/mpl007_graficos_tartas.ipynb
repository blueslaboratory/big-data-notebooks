```python
import matplotlib.pyplot as plt
```

Ejemplo sencillo

```python
fig, ax = plt.subplots()

animales = [2, 3, 15]
nombres = ['pájaros', 'gatos', 'perros']
resalte = [0.0, 0.2, 0.0] # que gatos en este caso esté un poco fuera

ax.pie(animales, resalte, labels = nombres, autopct='%d%%', shadow = True, startangle = 90)

fig.suptitle("Gráfico de tarta")

plt.show()
```

Comprobamos que no haya nulos en la clase del billete

```python
titanic.Pclass.isna().sum()
```

Una manera sencilla de contar el número de billetes de cada clase es con value_counts(). Por defecto, value_counts() devuelve una serie con el número de ocurrencias de cada valor, el índice indica de qué valor se trata.

```python
cuenta = titanic.Pclass.value_counts()
cuenta
```

```python
fig, ax = plt.subplots()

ax.pie(cuenta, labels = cuenta.index, autopct='%d%%')

fig.suptitle("Proporción de billetes de cada clase")

fig.legend()

plt.show()
```
