E002. Para el fichero "statage.csv", muestra una figure con dos subplots, uno para ni침os y otro para ni침as. En cada uno hay que poner altura mediana, percentil 25 y percentil 75 para cada edad.

```python
# los argumentos indican que habr치 dos filas y una columna de subplots
fig, ax = plt.subplots(nrows = 1, ncols = 2, sharey=True)


fig.suptitle("Figure con 2 subplots")


# Forma 1
# Plotear M, P25 y P75 para chicas en el primer subplot
ax[0].plot(chicas['Agemos'], chicas['M'], label="M", color='blue')
ax[0].plot(chicas['Agemos'], chicas['P25'], label="P25", color='green', linestyle='--')
ax[0].plot(chicas['Agemos'], chicas['P75'], label="P75", color='orange', linestyle='--')

ax[0].legend()
ax[0].set_title('Chicas')
ax[0].set_xlabel('Edad (meses)')
ax[0].set_ylabel('Altura (cm)')
ax[0].grid(True)



# Forma 2
# Plotear M, P25 y P75 para chicos en el segundo subplot

# el parametro 'Agemos' es data['Agemos'] siendo data=datos[datos.Sex==2][]
# el parametro 'M' es data['M'] siendo data=datos[datos.Sex==2][]
ax[1].plot('Agemos', 'M', data=datos[datos.Sex==2], label="Mediana", color='blue')
ax[1].plot('Agemos', 'P25', data=datos[datos.Sex==2], label="P25", color='green', linestyle='--')
ax[1].plot('Agemos', 'P75', data=datos[datos.Sex==2], label="P75", color='orange', linestyle='--')

ax[1].legend()
ax[1].set_title('Chicos')
ax[1].set_xlabel('Edad (meses)')
ax[0].set_ylabel('Altura (cm)')
ax[1].grid(True)


plt.show()
```
