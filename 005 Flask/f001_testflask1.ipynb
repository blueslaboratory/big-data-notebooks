```python
import base64

from io import BytesIO
from flask import Flask
from matplotlib.figure import Figure

from pyngrok import ngrok

app = Flask(__name__)



@app.route("/")
def hello():

    # Generate the figure **without using pyplot**.
    fig = Figure()
    ax = fig.subplots()
    ax.plot([1, 2])

    # Save it to a temporary buffer.
    buf = BytesIO()
    fig.savefig(buf, format="png")

    # Embed the result in the html output.
    data = base64.b64encode(buf.getbuffer()).decode("ascii")

    return f"<img src='data:image/png;base64,{data}'/>"


# Configurar ngrok con tu token de autenticación
!ngrok authtoken "2pOQ7xnkIFyr8G4j5kPhMdeeYEy_39rHv8SLzwALSp1eCpTi3" # Reemplaza con tu token

# Exponer la aplicación usando ngrok
public_url = ngrok.connect(5000)
print(f"Tu aplicación Flask está disponible públicamente en: {public_url}")

app.run(port=5000)
```
