```python
import plotly.express as px
df = px.data.medals_long()

fig = px.bar(df,
             x="nation",
             y="count",
             color="medal",
             pattern_shape="medal",
             pattern_shape_sequence=[".", "x", "+"])
fig.show()
```

```python
print(df.columns)
```

```python
grouped = df.groupby(['nation', 'medal']).sum().reset_index().sort_values(by=['nation', 'count'], ascending=False)
grouped
```

```python
import plotly.express as px
df = px.data.carshare()
fig = px.scatter_mapbox(df,
                        lat="centroid_lat",
                        lon="centroid_lon",
                        color="peak_hour",
                        size="car_hours",
                        color_continuous_scale=px.colors.cyclical.IceFire,
                        size_max=15,
                        zoom=10)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
