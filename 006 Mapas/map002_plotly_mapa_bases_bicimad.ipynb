```python
# path = ../datos/
path = "/content/drive/My Drive/Clara del Rey/BIG DATA/Programacion/FTP/000 Datos/"
bases = pd.read_csv(path + "bases_bicimad.csv", sep = ";")
bases.head()
```

```python
fig = px.scatter_mapbox(bases,
                        lat="Latitud",
                        lon="Longitud",
                        hover_name="Direccion",
                        color='Número de Plazas',
                        size='Número de Plazas',
                        zoom=11,
                        height=500)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
