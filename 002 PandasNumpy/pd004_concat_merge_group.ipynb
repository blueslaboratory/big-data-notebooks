Creamos un dataframe con marcas de coches y sus países de origen



```python
import pandas as pd
import numpy as np
# numpy: NUMerical_PYthon

nombremarcas1 = ['Renault','Toyota','Ford']
paises1 = ["Francia","Japón", "USA"]

marcas1 = pd.DataFrame(nombremarcas1, columns = ['Marca'])
marcas1
```

Otro igual con otros datos

```python
marcas1["Pais"] = paises1
marcas1
```

```python
nombremarcas2 = ['Citroen','Nissan','GM']
paises2 = ["Francia","Japón", "USA"]

marcas2 = pd.DataFrame(nombremarcas2, columns = ['Marca'])
marcas2
```

```python
marcas2["Pais"] = paises2
marcas2
```

Concatenar horizontalmente (axis = 1): unirlos en uno que tenga las 4 columnas:

```python
marcas = pd.concat([marcas1, marcas2], axis = 1, ignore_index= True)
marcas
```

Concatenar verticalmente (axis = 0): unirlos en uno que tenga las 6 filas:

(por defecto)


```python
marcas = pd.concat([marcas1, marcas2], axis = 0, ignore_index= True)
marcas
```

Concatenar horizontalmente: unirlos en uno que tenga las 4 columnas: axis = 1


```python
numemple = [10000, 15000, 20000]
sede = ['París', 'Tokio', 'Detroit']

numempleDf =  pd.DataFrame(numemple, columns = ['numEmple'])
numempleDf['sede'] = sede

numempleDf
```

```python
marcosConEmple = pd.concat([marcas1, numempleDf], axis = 1)
marcosConEmple
```

## Merge
Empezamos creando un nuevo dataframe con algunos datos en común con el anterior


```python
empleados_dict = {
    'Marca': ['Renault', 'Toyota', 'Ford','Citroen','Subaru', 'Skoda'],
    'Empleados' :[10000, 15000, 20000, 25000, 30000, 40000]
}

empleados = pd.DataFrame(empleados_dict)
empleados
```

Para unirlos en un solo dataframe con las tres columnas, se puede usar merge()

```python
pd.merge(marcas, empleados)

#equivalentes
#pd.merge(marcas, empleados, on = "Marca")
#pd.merge(marcas, empleados, left_on = "Marca", right_on = "Marca")
```

```python
pd.merge(marcas, empleados, how='left')
```

```python
pd.merge(marcas, empleados, how='right')
```

```python
pd.merge(marcas, empleados, how='outer')
```

```python
# producto cartesiano:
pd.merge(marcas, empleados, how='cross').head(15)
```

Para probar el merge por índice, añadimos un índice a los dataframes, va a ser un supuesto código de marca


```python
marcascodigo = marcas.copy()

# inventamos unos codigos para la tabla de marcas
codigos = ['c1','c2','c3','c4','c5','c6']

marcascodigo.index = codigos

marcascodigo
```

Lo mismo con el dataframe de empleados, vigilando que los códigos cuadren

```python
empleadoscodigo = empleados.copy()

codigos = ['c1','c2','c3','c4','c7','c8']

empleadoscodigo.index = codigos

empleadoscodigo
```

```python
# Combinar 2 DataFrames: marcascodigo & empleadoscodigo
# utilizando sus índices como clave de combinación: left_index=True, right_index=True
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html#pandas.DataFrame.merge

pd.merge(marcascodigo, empleadoscodigo, left_index=True, right_index=True)
```

Como alternativa, muy similar, se puede usar el método join

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.join.html#pandas.DataFrame.join

## Agrupar

```python
tabla = pd.merge(marcascodigo, empleadoscodigo, left_index=True, right_index=True)
tabla
```

```python
tabla.drop(columns="Marca_y", inplace=True)
tabla
```

```python
tabla.rename(columns={"Marca_x": "Marca"}, inplace=True)
tabla
```

Por defecto, groupby devuelve los datos en una serie indexada con los grupos.

```python
# Cuantas marcas hay de cada pais
numMarcas = tabla.groupby("Pais")["Marca"].count()

numMarcas
```

```python
numMarcas2 = tabla.groupby("Pais")["Marca"].unique()

numMarcas2
```

```python
numMarcas['USA']
```

```python
numEmple = tabla.groupby("Pais")["Empleados"].sum()
numEmple
```

Con as_index=False devuelve un dataframe

```python
numMarcas2 = tabla.groupby("Pais",as_index=False)["Marca"].count()
numMarcas2
```

```python
tabla2 = tabla.copy()
tabla2['Ventas'] = np.linspace(100000, 200000, 4)
tabla2
```

Para calcular varias funciones de grupo, se usa aggregate

```python
numMarcas2 = tabla2.groupby("Pais")["Empleados"].aggregate(["sum", "mean"])
numMarcas2
```

Si el resultado de la agrupación, tiene más de una columna, se crean varios niveles

```python
numMarcas2 = tabla2.groupby("Pais")[["Empleados", "Ventas"]].aggregate(["sum", "mean"])
numMarcas2
```

Para seleccionar uno de ellos

```python
numMarcas2["Empleados"]
```

Y para seleccionar uno de los grupos, como el nombre del grupo en este caso está en el índice, usamos loc

```python
numMarcas2["Empleados"].loc["Francia"]
```

Para iterar sobre los grupos

```python
# cada uno de los grupos tiene un nombre, que es el nombre del pais
grupos = tabla.groupby("Pais")
grupos.head(5)
```

```python
for pais, datos in grupos:
  print(pais)
  print(datos.head())
```

Para obtener directamente uno de los grupos como un dataframe

```python
grupos.get_group("Francia")
```
