# Media vs Mediana
Date: 10/10/2024

```python
import pandas as pd
import numpy as np
```

```python
# esto es un array con los numeros del 1-100: [0, 99]
a = np.linspace(1, 100, 100)
a[99] = 1000
```

```python
print(a)
```

```python
df = pd.DataFrame(a, columns = ["valor"])
df.head(4)
```

```python
df.valor.mean()
```

```python
# Conjuntos con un n√∫mero par de elementos (100 elementos):
# La mediana se calcula como el promedio de los dos valores centrales.
# 50 + 51 / 2 = 50.5

df.median()
```

```python
# si cambiamos 1000 por 1000000

# soon to be deprecated in pandas 3.0
df['valor'][99] = 1000000

# not soon to be deprecated
df.loc[99, "valor"] = 1000000
```

```python
print(f"Media: {df.valor.mean()}")
print(f"Mediana 1: {df.iloc[:, 0].median()}")
print(f"Mediana 2: {df.median()[0]}")
```
