A001. A partir del fichero netflix_titles.csv calcula el número de películas de cada país, su duración máxima, mínima y media.

```python
# para google colab:
# path = "../datos/"
path = "/content/drive/My Drive/Clara del Rey/BIG DATA/Programacion/FTP/000 Datos/"
datos = pd.read_csv(path + "netflix_titles.csv", sep = ",")
```

```python
datos.head(3)
```

```python
datos.describe()
```

```python
# peliculas
pelis = datos[datos.type == "Movie"].copy()
len(pelis)
```

```python
# pelis.duration.fillna("0 min", inplace = True)
pelis.fillna({"duration": "0 min"}, inplace = True)

pelis.duration.str[:-4]
```

```python
pelis.head(5)
```

```python
# Asegúrate de que la columna de duración esté en formato numérico, extrayendo los números de la cadena
pelis['duration'] = pelis['duration'].str.replace(' min', '').astype(int)

# Agrupar por país y calcular el número de películas, duración máxima, mínima y media
resultado = pelis.groupby('country')['duration'].agg(
    num_peliculas='count',
    duracion_max='max',
    duracion_min='min',
    duracion_media='mean'
)

# Mostrar el resultado
resultado
```

A002. Guarda esta información en un fichero .csv con con columnas PAIS, NUMERO, MAXIMA, MINIMA Y MEDIA

```python
# Definir la ruta donde se guardará el archivo
ruta = path + "pd007_netflix.csv"

# Restablecer el índice para que 'country' sea una columna normal y poder hacer el siguiente paso
resultado = resultado.reset_index()

# Renombrar las columnas del DataFrame resultado, sin el reset_index(), daría una columna menos
resultado.columns = ['PAIS', 'NUMERO PELICULAS', 'MAXIMA', 'MINIMA', 'MEDIA']

# Guardar el DataFrame en el archivo CSV en la ruta especificada
resultado.to_csv(ruta, index=False)

print("El archivo ha sido guardado correctamente en:")
print(ruta)
```
