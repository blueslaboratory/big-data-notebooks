Para seleccionar a los que no tienen edad nula

```python
# ~: es un operador de negacion
edadnonula = datos[~datos.Age.isnull()]
len(edadnonula)
```

Para rellenar los valores nulos

```python
# inplace = True --> modifica el dataframe datos.Age con datos.Age.mean()
datos.Age.fillna(datos.Age.mean(), inplace = True)

# inplace = False (por defecto) --> habria que asignarlo, devuelve una serie con los valores reemplazados
datos.Age = datos.Age.fillna(datos.Age.mean(), inplace = False)
```
