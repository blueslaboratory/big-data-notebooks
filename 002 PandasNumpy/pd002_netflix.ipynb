A003. Cuenta el número de directores diferentes que aparecen en el fichero.

```python
datos.shape
```

```python
# Forma 1
# indice y la fila:
directors = []

for index, row in datos.iterrows():
    if row['director'] not in directors:
        directors.append(row['director'])

print(len(directors))
```

A004. Añade una nueva columna, "peliLarga", con valor 1 para las películas de más de 120 minutos y 0 en otro caso.

```python
# Forma 0: (correccion profesor)
datos.duration
```

```python
pelis = datos[datos.type=="Movie"].copy()
pelis.fillna({"duration": "0 min"}, inplace = True)
pelis.duration.str[:-4]
```

```python
peliLarga = pelis.duration.str[:-4].astype(int) > 120
peliLarga
```

```python
pelis["peliLarga"] = peliLarga
pelis
```

```python
# Forma 1:

def is_long_movie(duration):
    if isinstance(duration, str):
        if 'min' in duration:
            minutes = int(duration.split()[0])
            if minutes >= 120:
                return 1
            else:
                return 0
        elif 'Season' in duration:
            return 0
    return 0
```

```python
peliLarga = movies.duration.apply(is_long_movie)
peliLarga
```

```python
movies['peliLarga'] = peliLarga
movies.head()
```

```python
# Forma 2:
# isinstance(duration, str): Esto asegura que duration sea una cadena antes de intentar buscar "min" o dividirla.

movies['peliLarga'] = [
    1 if isinstance(duration, str) and 'min' in duration and int(duration.split()[0]) >= 120 else 0
    for duration in movies['duration']
]

movies.head()
```
