```python
# iloc[todas las filas, desde la columna 2]
# Sumar todas las columnas desde la columna 2 (incluyendo las posteriores) del DataFrame 'latina',
# y solo considerar las columnas numéricas para la suma.
suma_latina = latina.iloc[:, 2:].sum(numeric_only=True)
suma_latina
```

```python
# Crear un nuevo DataFrame a partir de la suma calculada, transponiéndolo para que las filas se conviertan en columnas.
df_suma_latina = pd.DataFrame(suma_latina).T
df_suma_latina
```

```python
# Insertar una nueva columna 'DISTRITO' en la primera posición del DataFrame,
# con el valor constante 'LATINA' para todas las filas.
df_suma_latina.insert(0, 'DISTRITO', 'LATINA')

# Insertar otra columna 'AÑO' en la primera posición del DataFrame,
# con el valor constante 2023 para todas las filas.
df_suma_latina.insert(0, 'AÑO', 2023)

# Mostrar el DataFrame final que contiene la suma de las columnas, el distrito y el año.
df_suma_latina
```

A004. Encontrar el distrito con más daños por fuegos entre 2023 y 2024.

```python
# Crear una nueva columna con el valor "2023-2024" para unificar los años
bomberos_2023_24['AÑO'] = '2023-2024'
bomberos_2023_24.head(5)
```

```python
# Agrupar por DISTRITO y MES, sumando todas las columnas numéricas
suma_salidas_2023_2024 = bomberos_2023_24.groupby(['DISTRITO', 'MES']).sum(numeric_only=True)
suma_salidas_2023_2024.head(5)
```

```python
# El método .reset_index() asegura que los valores de DISTRITO y MES queden como columnas normales en lugar de índices.
suma_salidas_2023_2024 = bomberos_2023_24.groupby(['DISTRITO', 'MES']).sum(numeric_only=True).reset_index()
suma_salidas_2023_2024.head(5)
```

```python
# Mostrar las primeras 25 filas del DataFrame resultante
suma_salidas_2023_2024.head(25)
```

```python
bomberos_2023_24['MES'] = 'sumatorio anual'

suma_salidas_anual = bomberos_2023_24.groupby(['DISTRITO', 'MES']).sum(numeric_only=True)

suma_salidas_anual.head(4)
```

```python
# Resumen: max_fuegos_fila_anual contiene toda la fila del DataFrame que tiene el máximo valor en 'FUEGOS'

# Obtener el índice de la fila con el valor máximo en la columna 'FUEGOS'
# 'FUEGOS' es una de las columnas del DataFrame suma_salidas_anual

# 1. Primero, accedemos a la columna 'FUEGOS' del DataFrame
# Esta variable contiene la columna 'FUEGOS'
columna_fuegos = suma_salidas_anual['FUEGOS']
columna_fuegos.head(3)
```

```python
# 2. Luego, encontramos el índice del valor máximo en esa columna
# idxmax() retorna el índice de la primera ocurrencia del valor máximo
indice_max_fuegos = columna_fuegos.idxmax()
indice_max_fuegos
```

```python
# 3. Finalmente, usamos loc para obtener la fila completa correspondiente a ese índice
# loc selecciona la fila del DataFrame usando el índice encontrado
max_fuegos_fila_anual = suma_salidas_anual.loc[indice_max_fuegos]
print(max_fuegos_fila_anual)
```

```python
# Mostrar el distrito y el valor máximo de FUEGOS
print(f"Distrito con valor máximo de fuegos: {max_fuegos_fila_anual.name[0]}")
print(f"Numero de fuegos: {max_fuegos_fila_anual['FUEGOS']} fuegos")
```
