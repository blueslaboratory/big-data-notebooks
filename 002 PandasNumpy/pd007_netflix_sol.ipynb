```python
# solo pelis
pelis = datos[datos.type=="Movie"].copy()

pelis.duration.fillna("0 min", inplace = True)
pelis.duration = pelis.duration.str[:-4].astype(int)

pelis.groupby("country")["duration"].aggregate(["count", "max","min","mean"])
```
