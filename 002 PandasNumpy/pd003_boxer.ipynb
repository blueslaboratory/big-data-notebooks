A001. Calcula el porcentaje de hombres y mujeres en el fichero.

```python
# numero de filas
n_total = len(datos)
n_total
```

```python
# numero de hombres
n_hombres = (datos["Sex"] == "Male").sum()
n_hombres
```

```python
hombres = datos[datos.Sex == "Male"]
len(hombres)
```

```python
# numero de mujeres
n_mujeres = len(datos[datos.Sex == "Female"])
n_mujeres
```

```python
datos.groupby(["Sex"]).Rating.count()
```

```python
percentage_hombres = round((n_hombres/n_total)*100,2)
percentage_hombres
```

```python
percentage_mujeres = round((n_mujeres/n_total)*100,2)
percentage_mujeres
```

A002. Crea un DataFrame solo con los que pesen más de 150.

```python
# Forma 0 (profe)
boxeadores = datos.copy()
```

```python
boxeadores.Weight.unique()
```

```python
boxeadores["nuevoPeso"] = [151 if x == "HWT" else x for x in boxeadores.Weight]
```

```python
# los valores que habia antes eran antes strings, asi que hay que pasarlos a int
boxeadores.nuevoPeso = boxeadores.nuevoPeso.astype(int)

boxeadores.nuevoPeso.unique()
```

```python
box_pesados = boxeadores[boxeadores.nuevoPeso > 150].copy()
box_pesados.drop(columns = "nuevoPeso", inplace = True)
```

```python
box_pesados.head()
```

```python
# Forma 1: no saca los HWT (>150)
pesados = datos[pd.to_numeric(datos['Weight'], errors='coerce') >= 150]
pesados.head()
```

```python
len(pesados)
```

Salvar ese dataframe a un fichero "pesados.csv"


https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html
