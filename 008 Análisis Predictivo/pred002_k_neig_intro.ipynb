```python
# 14/01/2024

from sklearn import datasets
import pandas as pd
import seaborn as sns
import numpy as np
```

Cargar los datos

```python
x, y = datasets.load_iris(as_frame=True, return_X_y=True)
```

```python
y.unique()
```

Cambiamos la etiqueta a 30 elementos, aleatoriamente, por una cuarta etiqueta

```python
cambio =  np.random.randint(0, len(x), 30)
cambio
```

```python
y.loc[cambio] = 3
```

¿A qué especie asignarías los puntos con etiqueta 3?

```python
sns.scatterplot(data = x, x = "sepal length (cm)", y ="sepal width (cm)", hue = y).set(title = "¿A qué especie asignarías los puntos con etiqueta 3?")
```
