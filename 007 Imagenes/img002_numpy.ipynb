Vamos a crar una imagen a partir de una matriz de numpy:
- 800 por 600
- Fondo negro
- Franja horizontal roja.
- Franja horizontal azul.

```python
#17/12/2024
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matriz = np.zeros((600,800,3))
plt.imshow(matriz)
plt.show()
```

```python
# para algunas filas, todas las columnas, el primer canal  al máximo
# 1er parametro (eje y) | 2º parametro (eje x) | 3º parametro (canal rojo): R(0)G(1)B(2)
matriz[100:220:2,:, 0] = 1.0

plt.imshow(matriz)
plt.show()
```

```python
# para algunas filas, todas las columnas, color azul
# 1er parametro (eje y) | 2º parametro (eje x) | 3º parametro (canal rojo): R(0)G(1)B(2)
matriz[500:550,:, 2] = 1.0

plt.imshow(matriz)
plt.show()
```

```python
# un pixel en la  franja roja
matriz[100,5]
```

Para bajar las barras

```python
matriz = np.roll(matriz, axis = 0, shift = 100)
```

```python
plt.imshow(matriz)
plt.show()
```

Coger un trozo de la imagen: por ejemplo, la esquina superior derecha (100 píxeles en horizontal y 200 en vertical)

```python
matriz[0:50, 0:20, 1] = 1.0
plt.imshow(matriz)
plt.show()
```
